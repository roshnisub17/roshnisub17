#1. Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
csv_file = 'Roshni_Lifestyle.csv'

# Read the CSV data into a DataFrame
df = pd.read_csv(csv_file)

# Extract data from the "Occupation" (Column B) and "Stress Level" (Column H) columns
x_data = df['Occupation']
y_data = df['Stress Level']

# Create a bar chart without annotations
plt.figure(figsize=(10, 6))
plt.bar(x_data, y_data, color='pink')
plt.xlabel('Occupation')
plt.ylabel('Stress Level')
plt.title('Stress Level by Occupation')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

The distribution of a single variable (in this example, "Stress Level") is shown using a histogram. It creates intervals or "bins" out of the data and shows how frequently or how many data points fall into each bin.
The histogram shows the distribution of stress levels among individuals. It helps us understand how stress levels are distributed across the dataset.
From the above histogram, we can see software engineer, doctor and sales representative having the highest stress level and Manager having the lowest stress level.

#2. Bar Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
csv_file = 'Roshni_Lifestyle.csv'

# Read the CSV data into a DataFrame
df = pd.read_csv(csv_file)

# Extract data from the "Occupation" (Column B) and "Stress Level" (Column H) columns
x_data = df['Occupation']
y_data = df['Stress Level']

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(x_data, y_data, color='pink')
plt.xlabel('Occupation')
plt.ylabel('Stress Level')
plt.title('Stress Level by Occupation')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

A bar chart, where each category is represented by a bar, is a popular approach to display categorical data. Each bar's height reflects the value connected to that category.

Comparing "Stress Level" across various "Occupation" categories is done using a bar chart. Each occupational group is depicted on the x-axis, and the height of the bars on the y-axis denotes the appropriate stress level.

From the above bar chart, we can see that Software Engineer, Doctor and Sales Representative have the highest stress level while Manager has the lowest stree level.

#3. Line Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
csv_file = 'Roshni_Lifestyle.csv'

# Read the CSV data into a DataFrame
df = pd.read_csv(csv_file)

# Filter the data for specific genders (e.g., 'Male' and 'Female')
selected_genders = ['Male', 'Female']
filtered_data = df[df['Gender'].isin(selected_genders)]

# Group the data by 'Occupation' and calculate the average 'Stress Level'
grouped_data = filtered_data.groupby(['Gender', 'Occupation'])['Stress Level'].mean().unstack()

# Create a line graph
plt.figure(figsize=(12, 6))

for gender in selected_genders:
    plt.plot(grouped_data.columns, grouped_data.loc[gender], marker='o', label=gender)

plt.xlabel('Occupation')
plt.ylabel('Average Stress Level')
plt.title('Average Stress Level by Occupation and Gender')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

An interval or time series of data points are shown using a line chart. It is appropriate for displaying data patterns and trends.
The line graph above illustrates the connection between "Occupation" and "Stress Level" of each gender. The line connecting the dots on the graphic, which each represent a profession, illustrates how stress levels vary between various professions.
From the above graph, we can see the trend of stress level of each gender in each occupation.

#4. Pie Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
csv_file = 'Roshni_Lifestyle.csv'

# Read the CSV data into a DataFrame
df = pd.read_csv(csv_file)

# Extract data from the "Gender" and "BMI Category" columns
gender_counts = df['Gender'].value_counts()
bmi_category_counts = df['BMI Category'].value_counts()

# Create a pie chart for Gender
plt.figure(figsize=(12, 5))

# Pie chart for Gender
plt.subplot(1, 2, 1)
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('Distribution of Gender')

# Pie chart for BMI Category
plt.subplot(1, 2, 2)
plt.pie(bmi_category_counts, labels=bmi_category_counts.index, autopct='%1.1f%%', colors=['lightgreen', 'lightyellow', 'lightpink'])
plt.title('Distribution of BMI Category')

plt.tight_layout()
plt.show()
```

A pie chart is used to show how categorical data is distributed like a "pie," with each category denoting a piece of the pie. Each slice is proportionate to the percentage of the total that each group makes up.
This pie chart shows the distribution of stress levels, labeled as "Low," "Medium," and "High." The proportion of individuals in each category of stress level as a percentage of all people is shown by each pie slice.

From the excel sheet, I wanted to know how gender is distributed and how the BMIcategory is distributed. Hence, from the above pie chart, we can see that the excel sheet contains 50.5% male and 49.5% female. It also shows how the BMI category is distributed, with "normal" being the highest percentage.

```python
plt.plot(data['mpg'], data['weight'], 'o')
plt.show()
```

```python
plt.scatter(data['mpg'], data['weight'])
plt.show()
```

##**Bar Graph**

```python
cylinders = data['cylinders'].unique()
print(cylinders)
print(len(cylinders))
```

```python
cy_8 = data['cylinders'].value_counts()[8]
cy_8
```

```python
fig, ax = plt.subplots()
labels = cylinders

cy_8 = data['cylinders'].value_counts()[8]
cy_4 = data['cylinders'].value_counts()[4]
cy_6 = data['cylinders'].value_counts()[6]
cy_3 = data['cylinders'].value_counts()[3]
cy_5 = data['cylinders'].value_counts()[5]

counts = [cy_8, cy_4, cy_6, cy_3, cy_5]
colors = ['red', 'blue', 'purple', 'orange', 'green']

ax.bar(labels, counts, label= labels, color=colors)

ax.set_ylabel('Count of Cylinders')
ax.set_title('Cylinders by Count')
```
